```python
import collections
class Country:
    def __init__(self, name, gdp):
        self.name = name
        tempDict = collections.OrderedDict()
        for key, val in gdp.items():
            if val != '':
                tempDict.update({key, float(val)})
        self.gdp = tempDict

    def printName(self):
        print(self.name)
    
    def getName(self):
        return(self.name)

    def getAllGDP(self):
        return self.gdp

    def getGDP(self, key):
        return self.gdp.get(key)

    def listYears(self):
        return self.gdp.keys()
    
    def getHighestChange(self):
        highest = 0.0
        for key in self.gdp:            
            if self.gdp[key] > highest:
                highest = self.gdp[key]
        return highest
    
    def getLowestChange(self):
        lowest = 0.0
        for key in self.gdp:            
            
            if self.gdp[key] < lowest:
                lowest = self.gdp[key]
        return lowest
    
   # def getRangeGDP(self, start, end):
       # temp = list()
       ## for key in self.gdp:
        
        #return list(k:adict[k] for k in ('key1','key2','key99') if k in adict)
```

```python
import csv
import collections
countryDict = dict()
#defining a variable quarters, which will be set to a list of the available quarters in the data table.
#Essentially, a list of all keys
quarters = None
temp = 0
with open('oecd-gdp-pc-change-1997-2017.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        name = row.pop('Country')
        if quarters == None:
            quarters = list(row.keys())
        countryDict[name] = Country(name, collections.OrderedDict(row))

#  Printing every value of every quarter for every country
#for c in countryDict:
#    for q in countryDict[c].getAllGDP():
#        print(c + ' ' + ' '+ q + ' ' + str(countryDict[c].getGDP(q)))
```

```python
#Q1  Find the country and quarter with the highest percentage change and that with
 #the lowest precentage change. Some countries are missing data for some quarters
# (the entry for that quarter is blank), so these missing quarters should be ignored.

##############################################
### Should probably define this section as a seperate function
### since it gets repeated in Q1 and Q2
##############################################
high = 0.0
low = 0.0
highestCountry = ''
lowestCountry = ''

for key in countryDict:
    if countryDict[key].getHighestChange() > high:
        high = countryDict[key].getHighestChange()
        highestCountry = countryDict[key].getName()
    if countryDict[key].getLowestChange() < low:
        low = countryDict[key].getLowestChange()
        lowestCountry = countryDict[key].getName()
##############################################

print(highestCountry + ' has the highest GDP with ' + str(high) + '%')
print(lowestCountry + ' has the lowest GDP with ' + str(low) + '%')
```

```python
#Q2 For each quarter, take the average of the changes in GDP across all countries
#(ignoring blank entries). Find the quarter with the highest average change and
#the quarter with the lowest average change.

#setting the average quarter change as a dictionary
avgQuarterChange = dict()

#Iterating through the quarters dictionary
for q in quarters:
    #setting a temporaty float
    temp = 0.0
    c = 0
    #iterating through the country dictionary
    for key in countryDict:
        #adding the GDP of the current quarter for the current country to the temporary value
        if countryDict[key].getGDP(q) != None:
            temp += countryDict[key].getGDP(q)
            c+=1
            #print (str(key) + ' ' + str(q) + ' ' + str(countryDict[key].getGDP(q)))
    #calculating the average change for the current quarter
    avgQuarterChange[q] = temp/c

##############################################
### Should probably define this section as a seperate function
### since it gets repeated in Q1 and Q2
##############################################
high = 0.0
low = 0.0
highestQ = ''
lowestQ = ''

for key in avgQuarterChange:
    if avgQuarterChange[key] > high:
        high = avgQuarterChange[key]
        highestQ = key
    if avgQuarterChange[key] < low:
        low = avgQuarterChange[key]
        lowestQ = key
##############################################
print('The highest average change was ' + str(high) + ' in ' + highestQ)
print('The lowest average change was ' + str(low) + ' in ' + lowestQ)

#Answers should be:
#Highest change: 1.584068647%, 1997-Q2
#Lowest change: -2.692734314%, 2009-Q1
```

```python
#Q3 Find the cumulative percentage change for each country. To do so, start with
#1 and for each quarter (that isn't blank) multiply it by (1 + change/100).
#Eg, the first two changes for Greece (GRC) are 0.575928 and 1.69888, so the
#cumulative change is 1 * (1 + 0.575928/100) * (1 + 1.69888/100) = 1.0228, or
#+2.28 %. Rank each country according to its cumulative change in GDP.

cumuChangeDict = dict()
#calculate the 
for key in countryDict:
    c = 1
    for v in countryDict[key].getAllGDP().values():
        c *= (1+v/100)
    cumuChangeDict[c] = key

rankedCumuChangeDict = dict()
for c in sorted(list(cumuChangeDict.keys()),reverse = True):
    rankedCumuChangeDict[cumuChangeDict[c]] = c
    
for key in rankedCumuChangeDict:
    print(key + ' ' + str(rankedCumuChangeDict[key]) + ' or ' + str((rankedCumuChangeDict[key]-1.0)*100.0)+'%')
```
